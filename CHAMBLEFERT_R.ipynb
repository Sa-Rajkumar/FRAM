#ESTIMATING BETA USING CAPM (DAILY)


```r
install.packages("quantmod")
library(quantmod)
require(quantmod)

NSE <- getSymbols("^NSEI", from = "2020-04-01", to = "2024-10-01", auto.assign = FALSE) #Get data
NSE
CHAMBLFERT <- getSymbols("CHAMBLFERT.NS", from = "2020-04-01", to = "2024-10-01", auto.assign = FALSE) #Get data
CHAMBLFERT

head(NSE, 5)
tail(NSE, 5)

Close <- cbind(Cl(NSE), Cl(CHAMBLFERT)) #Dataframe of closing prices
head(Close, 5)

Returns <- as.xts(tail(data.frame(Close), -1) / head(data.frame(Close), -1) - 1) #Calculate returns
head(Returns, 5)

# Run the regression model
regression <- lm(CHAMBLFERT.NS.Close ~ NSEI.Close, data = data.frame(Returns))

# Slope parameter = beta in CAPM model
summary(regression)
```

```r
# Install and load necessary packages
install.packages("readxl")
install.packages("quantmod")
install.packages("PerformanceAnalytics") # For CAPM functions
library(readxl)
library(quantmod)
library(PerformanceAnalytics)

# Read the T-bill yield data from an Excel file
# Replace "path/to/your/tbill_data.xlsx" with the actual path to your Excel file
tbill_data <- read_excel("/content/T-Bills_2024.xlsx")

# View the data to ensure it's loaded correctly
print(tbill_data)

# Calculate the average monthly yield (assuming 'Monthly Yield %' is the column name)
average_monthly_yield <- mean(tbill_data$`Monthly Yield %`) / 100 # Convert percentage to decimal

# Annualize the risk-free rate (assuming monthly compounding)
annualized_risk_free_rate <- (1 + average_monthly_yield)^12 - 1

# Get market index data (NSE)
NSE <- getSymbols("^NSEI", from = "2020-04-01", to = "2024-10-01", auto.assign = FALSE)

# Get stock data (CONFIPET)
CONFIPET <- getSymbols("CONFIPET.NS", from = "2020-04-01", to = "2024-10-01", auto.assign = FALSE)

# Combine closing prices into a single dataframe
Close <- cbind(Cl(NSE), Cl(CONFIPET))

# Calculate returns
Returns <- as.xts(tail(data.frame(Close), -1) / head(data.frame(Close), -1) - 1)

# Run the regression model
regression <- lm(CONFIPET.NS.Close ~ NSEI.Close, data = data.frame(Returns))

# Display regression summary to find beta
beta_summary <- summary(regression)
beta_value <- coef(beta_summary)[2] # Extract beta value

# Expected market return (assumed; can be adjusted based on historical data)
expected_market_return <- mean(Returns[,1]) + annualized_risk_free_rate # Adjusted for risk-free rate

# CAPM formula: Expected Return = Risk-Free Rate + Beta * (Expected Market Return - Risk-Free Rate)
expected_return <- annualized_risk_free_rate + beta_value * (expected_market_return - annualized_risk_free_rate)

# Output results
cat("Beta of CONFIPET:", beta_value, "\n")
cat("Annualized Risk-Free Rate:", annualized_risk_free_rate, "\n")
cat("Expected Return using CAPM:", expected_return, "\n")
```

#ARIMA MODELING (DAILY)

```r
install.packages("tseries")
install.packages("ggplot2")
library(tseries)
library(ggplot2)
library(quantmod)

CHAMBLFERT <- getSymbols("CHAMBLFERT.NS", from = "2020-04-01", to = "2024-10-01", auto.assign = FALSE) #Get data
head(CHAMBLFERT, 5)

Returns_CHAMBLFERT <- as.xts(tail(data.frame(CHAMBLFERT$CHAMBLFERT.NS.Close), -1)/head(data.frame(CHAMBLFERT$CHAMBLFERT.NS.Close), -1) -1, frequency = 365) #Calculate returns
colnames(Returns_CHAMBLFERT) <- "Returns"
head(Returns_CHAMBLFERT)
plot(Cl(CHAMBLFERT), main = "CHAMBLFERT Closing Prices", ylab = "Price", col = "blue")
plot(Returns_CHAMBLFERT, main = "CHAMBLFERT Returns", ylab = "Returns", col = "blue")

# Identifying the Model using ADF test
adf_test_result <- adf.test(Returns_CHAMBLFERT$Returns, alternative = "stationary")
print(adf_test_result)

# ACF and PACF plots
plot(acf(Returns_CHAMBLFERT, lag.max = 10, main = "ACF of CHAMBLFERT Returns"))  # For AR terms
plot(pacf(Returns_CHAMBLFERT, lag.max = 10, main = "PACF of CHAMBLFERT Returns"))  # For MA terms

# Fitting an ARIMA model
arima_final <- arima(Returns_CHAMBLFERT$Returns, order = c(0, 0, 1))

# Prediction using this model
predicted <- predict(arima_final, n.ahead = 10)
print(predicted)

# Diagnostics of the model
tsdiag(arima_final)
```

#GARCH AND EGARCH MODELING (DAILY)

```r
install.packages("rugarch")
install.packages("quantmod")
library(quantmod)
library(rugarch)

CHAMBLFERT <- getSymbols("CHAMBLFERT.NS", from = "2020-04-01", to = "2024-10-01", auto.assign = FALSE) #Get data
head(CHAMBLFERT)

rCHAMBLFERT <- dailyReturn(CHAMBLFERT) #Calculate Daily Returns
head(rCHAMBLFERT)

# Implementing Univariate GARCH specification
ug_spec <- ugarchspec()
ug_spec

# Implementing EGARCH specification
eg_spec <- ugarchspec(variance.model = list(model = "eGARCH"))
eg_spec

# Estimating the GARCH model
ugfit <- ugarchfit(spec = ug_spec, data = rCHAMBLFERT)
ugfit

# Forecasting using the fitted GARCH model
ugforecast <- ugarchforecast(ugfit, n.ahead = 10)
ugforecast
```
